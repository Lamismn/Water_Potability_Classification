```python
# water_par[(water_par['DO']>=6) & (water_par['pH'] < 8.5) & (water_par['pH']>6.5) & (water_par['BOD']<2) & (water_par['TOTAL_COLIFORM']<50)]
A
def grade(water_par):
    if water_par['DO'] >= 6 and water_par['pH']<8.5 and water_par['pH']>6.5 and water_par['BOD']<2 and water_par['TOTAL_COLIFORM']<50:
        return 'A'
    elif water_par['DO'] >= 5 and water_par['pH']<8.5 and water_par['pH']>6.5 and water_par['BOD']<3 and water_par['TOTAL_COLIFORM']<500:
        return 'B'
    elif water_par['DO'] >= 4 and water_par['pH']<9 and water_par['pH']>6 and water_par['BOD']<3 and water_par['TOTAL_COLIFORM']<5000:
        return 'C'
    elif water_par['pH']<8.5 and water_par['pH']>6.5 and water_par['DO'] >= 4 and water_par['NITRATE_N_NITRITE_N']<1.2 and water_par['TOTAL_COLIFORM']<50000:
        return 'D'
    elif water_par['pH']<8.5 and water_par['pH']>6 and water_par['CONDUCTIVITY'] < 2250 and water_par['NITRATE_N_NITRITE_N']<2 and water_par['TOTAL_COLIFORM']<60000:
        return 'E'
    else:
        return 'F'
   

water_par['Grade'] = water_par.apply(grade, axis = 1)

water_par
```

```python
water_par.to_csv('Resources/water_grade.csv')
```

```python
def water_pot(water_par):
    if water_par['Grade'] == 'A' or water_par['Grade'] == 'C':
        return 1
    else:
        return 0
    
water_par['Potability'] = water_par.apply(water_pot, axis = 1)

water_par
```
