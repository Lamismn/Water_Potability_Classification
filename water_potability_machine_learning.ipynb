```python
#import dependancies and test data for pre-proccessing and scaling
import pandas as pd
file_load = "Resources/water_potability_cleaned.csv"
water_df = pd.read_csv(file_load)
water_df.drop("Unnamed: 0", axis = 1, inplace=True)
water_df.head()
```

```python
#Identifying our inputs and target for our machine learning model
inputs = ["ph", "Hardness", "Solids", "Chloramines", "Sulfate", "Conductivity", "Organic_carbon", "Trihalomethanes", "Turbidity"]

target = ["Potability"]
```

```python
#Seperating our values into inputs and targets
X = water_df.copy()
X.drop(target, axis = 1, inplace = True)

y = water_df[target]
```

```python
#did it work?
X.head()
```

```python
#Split the data into testing and training for our model
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
#Scaling our data so outliers do not affect our model too greatly
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaler = scaler.fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs',
                                max_iter=200,
                                random_state=1)
```

```python
classifier.fit(X_train_scaled, y_train.values.ravel())
```

```python
y_pred = classifier.predict(X_test_scaled)
results = pd.DataFrame({"Prediction": y_pred, "Actual": y_test.values.ravel()}).reset_index(drop=True)
results.head(20)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test.values.ravel(), y_pred))
```
